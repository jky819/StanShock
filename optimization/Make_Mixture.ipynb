```python
import sys; sys.path.append('../')
from stanShock import dSFdx, stanShock, smoothingFunction
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import time
import cantera as ct
from scipy.optimize import newton
```

```python
# parameters for shock ID 001
fontsize = 12
tFinal = 25e-3
p5, p1 = 97689.238746, 6906.898597
T5 = 988.497976
#g4 = g1 = 5.0/3.0 #monatomic gas in driver and driven sections
g4, g1 = 0.7044*(7/5)+0.2956*(5/3), 5/3
W4, W1 = 28*0.7044+4*0.2956, 39.948
MachReduction = 0.981 #account for shock wave attenuation
nXCoarse, nXFine = 200, 500 #mesh resolution
LDriver, LDriven = 3.63, 9.73
DDriver, DDriven = 1.4*(11.53e-2), 11.53e-2
plt.close("all")
mpl.rcParams['font.size']=fontsize
plt.rc('text',usetex=True)
```

```python
xLower = -LDriver
xUpper = LDriven
xShock = 0.0
Delta = 10*(xUpper-xLower)/float(nXFine)
geometry=(nXCoarse,xLower,xUpper,xShock)
DInner = lambda x: np.zeros_like(x)
dDInnerdx = lambda x: np.zeros_like(x)
def DOuter(x): return smoothingFunction(x,xShock,Delta,DDriver,DDriven)
def dDOuterdx(x): return dSFdx(x,xShock,Delta,DDriver,DDriven)
A = lambda x: np.pi/4.0*(DOuter(x)**2.0-DInner(x)**2.0)
dAdx = lambda x: np.pi/2.0*(DOuter(x)*dDOuterdx(x)-DInner(x)*dDInnerdx(x))
dlnAdx = lambda x,t: dAdx(x)/A(x)
```

```python
xLower = -LDriver
xUpper = LDriven
xShock = 0.0
Delta = 10*(xUpper-xLower)/float(nXFine)
geometry=(nXCoarse,xLower,xUpper,xShock)
# define discrete insert input
insert_loc = -0.2
def DInner(x):
    DInner_cur = np.zeros(x.shape)
    cnt = 0
    for X in x:
        if X <=insert_loc-0.6604:#2.1568:
            DInner_cur[cnt] = 0.0762
        elif X >insert_loc-0.6604 and X <= insert_loc:#2.1568 and X <= -1.4964:
            DInner_cur[cnt] = (9/208)*(insert_loc-X)+0.047625
        elif X > insert_loc:#-1.4964:
            DInner_cur[cnt] = (0.047625/Delta)*(insert_loc+Delta-X)
        cnt = cnt+1
    DInner_cur[x>insert_loc+Delta] = np.zeros_like(x[x>insert_loc+Delta])
    return  DInner_cur
x = np.linspace(-3.63, 0, 200)
D = DInner(x)
plt.plot(x, D)
plt.axis('equal')
def dDInnerdx(x):
    dDInnerdx_cur = np.zeros(x.shape)
    cnt = 0
    for X in x:
        if X <=insert_loc-0.6604:#2.1568:
            dDInnerdx_cur[cnt] = 0
        elif X >insert_loc-0.6604 and X <= insert_loc:#2.1568 and X <= -1.4964:
            dDInnerdx_cur[cnt] = -(9/208)
        elif X > insert_loc:#-1.4964:
            dDInnerdx_cur[cnt] = 0.047625/Delta
        cnt = cnt+1
    dDInnerdx_cur[x>insert_loc+Delta] = np.zeros_like(x[x>insert_loc+Delta])
    return dDInnerdx_cur
#x = np.linspace(-3.63, 0, 200)
#dD = dDInnerdx(x)
#plt.plot(x, dD)

def DOuter(x): return smoothingFunction(x,xShock,Delta,DDriver,DDriven)
def dDOuterdx(x): return dSFdx(x,xShock,Delta,DDriver,DDriven)
A = lambda x: np.pi/4.0*(DOuter(x)**2.0-DInner(x)**2.0)
dAdx = lambda x: np.pi/2.0*(DOuter(x)*dDOuterdx(x)-DInner(x)*dDInnerdx(x))
dlnAdx = lambda x,t: dAdx(x)/A(x)
```

```python
def res(Ms1):
    return p5/p1-((2.0*g1*Ms1**2.0-(g1-1.0))/(g1+1.0))\
                *((-2.0*(g1-1.0)+Ms1**2.0*(3.0*g1-1.0))/(2.0+Ms1**2.0*(g1-1.0)))
Ms1 = newton(res,2.0)
Ms1*= MachReduction
T5oT1 = (2.0*(g1-1.0)*Ms1**2.0+3.0-g1)\
       *((3.0*g1-1.0)*Ms1**2.0-2.0*(g1-1.0))\
       /((g1+1.0)**2.0*Ms1**2.0)
T1 = T5/T5oT1
a1oa4 = np.sqrt(W4/W1)
p4op1 = (1.0+2.0*g1/(g1+1.0)*(Ms1**2.0-1.0))\
       *(1.0-(g4-1.0)/(g4+1.0)*a1oa4*(Ms1-1.0/Ms1))**(-2.0*g4/(g4-1.0))
p4 = p1*p4op1

#set up the gasses
u1 = 0.0;
u4 = 0.0; #initially 0 velocity
mech = "Propane.xml" #mech="HeliumArgon.xml"
gas1 = ct.Solution(mech)
gas4 = ct.Solution(mech)
T4 = T1; #assumed
gas1.TPX = T1, p1,"AR:1"#gas1.TPX = T1,p1,"AR:1"
gas4.TPX = T4,p4, "N2:0.7044, HE:0.2956"#"N2:1"#gas4.TPX = T4,p4,"HE:1"
```

```python
boundaryConditions=['reflecting','reflecting']
state1 = (gas1,u1)
state4 = (gas4,u4)
```

```python
geometry=(nXFine,xLower,xUpper,xShock)
#gas1.TPX = T1, p1,"C3H8:0.0309, O2:0.195, HE:0.388, N2:0.3861"#gas1.TPX = T1,p1,"AR:1"
#gas4.TPX = T4,p4, "N2:0.8286, HE:0.1714"#"N2:1"#gas4.TPX = T4,p4,"HE:1"
gas1.TPX = T1, p1,"AR:1"#gas1.TPX = T1,p1,"AR:1"
gas4.TPX = T4,p4, "N2:0.7044, HE:0.2956"#"N2:1"#gas4.TPX = T4,p4,"HE:1"
ss = stanShock(gas1,initializeRiemannProblem=(state4,state1,geometry),
                   boundaryConditions=boundaryConditions,
                   cfl=.9,
                   outputEvery=100,
                   includeBoundaryLayerTerms=True,
                   Tw=T1, #assume wall temperature is in thermal eq. with gas
                   DOuter= DOuter,
                   DInner= DInner,
                   dlnAdx=dlnAdx)
ss.addXTDiagram("p")
ss.addXTDiagram("T")
ss.addProbe(max(ss.x)) #end wall probe
t0 = time.clock()
ss.advanceSimulation(tFinal)
t1 = time.clock()
print("The process took ", t1-t0)
pInsert = np.array(ss.probes[0].p)
tInsert = np.array(ss.probes[0].t)

ss.plotXTDiagram(ss.XTDiagrams["t"],limits=[200.0,700.0])


ss.plotXTDiagram(ss.XTDiagrams["p"],limits=[0.1,2])
xInsert = ss.x
DOuterInsert = ss.DOuter(ss.x)
DInnerInsert = ss.DInner(ss.x)
```
